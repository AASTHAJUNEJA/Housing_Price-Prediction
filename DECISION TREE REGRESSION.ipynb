### DECISION TREE IMPLEMENTATION 

### USING SINGLE FEATURE = SQFT_LIVING AND Y = PRICE USING DATA OF 4994 ROWS 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df4 = pd.read_csv("housingnew.csv")
```

```python
df4.head()
```

```python
y = np.array(df4["price"]).reshape(-1,1)
y.shape
```

```python
x = np.array(df4["sqft_living"]).reshape(-1,1)
x.shape
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor(random_state = 0) 
regressor.fit(x, y) 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=7)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
tree.plot_tree(regressor.fit(x_train,y_train))
```

```python
regressor.predict([[1000]])
```

### TAKING NUMBER OF ROWS AS 100 AND SINGLE FEATURE AS SQFT_LIVING

```python
x.columns
```

```python
x.shape
```

```python
x.describe()
```

```python
y.shape
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor(random_state = 0) 
regressor.fit(x, y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=7)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
tree.plot_tree(regressor.fit(x_train,y_train))
```
